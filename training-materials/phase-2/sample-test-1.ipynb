## 1. Variables and Data Types

```python
# Basic variable assignment
name = "Alice"
age = 30
score = 95.5
is_active = True

print(f"Name: {name}")
print(f"Age: {age}")
print(f"Score: {score}")
print(f"Active: {is_active}")
```

## 2. Lists and Loops

```python
fruits = ["apple", "banana", "cherry", "date"]

for fruit in fruits:
    print(f"Fruit: {fruit}")
```

## 3. Simple Function

```python
def greet(name):
    return f"Hello, {name}!"

print(greet("World"))
print(greet("Learner"))
```

## 4. Basic Arithmetic

```python
a = 12
b = 5

print(f"{a} + {b} = {a + b}")
print(f"{a} - {b} = {a - b}")
print(f"{a} * {b} = {a * b}")
print(f"{a} / {b} = {a / b:.2f}")
print(f"{a} % {b} = {a % b}")
```
